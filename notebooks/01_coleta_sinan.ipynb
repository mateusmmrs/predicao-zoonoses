# Coleta de Dados do SINAN
Download dos registros individuais de doenças de notificação compulsória: Leptospirose e Leishmaniose


```
# pip install pysus
from pysus.ftp.databases.sinan import SINAN

sinan = SINAN().load()

# Listar doenças disponíveis
print(sinan.get_files('LEPT'))

# Download de leptospirose, anos 2010-2023
for year in range(2010, 2024):
    try:
        files = sinan.get_files('LEPT', year)
        df = sinan.download(files).to_dataframe()
        df.to_parquet(f'../data/raw/sinan/leptospirose_{year}.parquet')
        print(f"  ✅ {year}: {len(df)} registros")
    except Exception as e:
        print(f"  ❌ {year}: {e}")

# Repetir para LEIV (Leishmaniose Visceral)
for year in range(2010, 2024):
    try:
        files = sinan.get_files('LEIV', year)
        df = sinan.download(files).to_dataframe()
        df.to_parquet(f'../data/raw/sinan/leish_visceral_{year}.parquet')
        print(f"  ✅ LEIV {year}: {len(df)} registros")
    except Exception as e:
        print(f"  ❌ LEIV {year}: {e}")

# LTAN (Leishmaniose Tegumentar)
for year in range(2010, 2024):
    try:
        files = sinan.get_files('LTAN', year)
        df = sinan.download(files).to_dataframe()
        df.to_parquet(f'../data/raw/sinan/leish_tegumentar_{year}.parquet')
        print(f"  ✅ LTAN {year}: {len(df)} registros")
    except Exception as e:
        print(f"  ❌ LTAN {year}: {e}")
```
